```python
import seaborn as sb
df = sb.load_dataset('tips')
df = df.groupby(['size', 'sex']).agg(mean_total_bill=("total_bill", 'mean'))
df = df.reset_index()
sb.barplot(x="size", y="mean_total_bill", hue="sex", data=df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dummy data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
subgroup1 = [20, 30, 40, 25, 35]
subgroup2 = [15, 25, 35, 30, 20]
subgroup3 = [10, 20, 30, 35, 25]

bar_width = 0.25

bar_positions1 = np.arange(len(categories))
bar_positions2 = [x + bar_width for x in bar_positions1]
bar_positions3 = [x + bar_width for x in bar_positions2]

plt.bar(bar_positions1, subgroup1, width=bar_width, label='Subgroup 1', color='red')
plt.bar(bar_positions2, subgroup2, width=bar_width, label='Subgroup 2', color='blue')
plt.bar(bar_positions3, subgroup3, width=bar_width, label='Subgroup 3', color='green')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Example of Grouped Bar Graph with Dummy Data')
plt.xticks([r + bar_width for r in range(len(categories))], categories)
plt.legend()

plt.show()
```
